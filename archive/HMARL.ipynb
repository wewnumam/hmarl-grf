```python
import numpy as np
import matplotlib.pyplot as plt

metrics = [
    "Chance prevention", "Intensity", "High line",
    "Deep build-up", "Press resistance", "Possession",
    "Central progression", "Circulate", "Field tilt",
    "Chance creation", "Shot quality", "Patient attack"
]

values = np.array([
    26, 31, 63,     # Defence
    90, 59, 54,     # Possession
    87, 65, 32,      # Progression
    68, 98, 39     # Attack
])

groups = (
    ["Defence"] * 3 +
    ["Possession"] * 3 +
    ["Progression"] * 3 +
    ["Attack"] * 3
)

colors = {
    "Defence": "#8C2D3E",
    "Possession": "#0F5B5B",
    "Progression": "#F2A31B",
    "Attack": "#4C78A8"
}

N = len(values)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False)
width = 2 * np.pi / N

fig, ax = plt.subplots(figsize=(9, 9), subplot_kw=dict(polar=True))
fig.patch.set_facecolor("#F7F7F4")
ax.set_facecolor("#F7F7F4")

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

# Bars
for i in range(N):
    ax.bar(
        angles[i],
        values[i],
        width=width * 0.95,
        color=colors[groups[i]],
        edgecolor="white",
        linewidth=1,
        align="edge"
    )

    # Value labels
    ax.text(
        angles[i] + width / 2,
        values[i] - 8,
        f"{values[i]}",
        ha="center",
        va="center",
        color="white",
        fontsize=11,
        fontweight="bold"
    )

ax.set_ylim(0, 100)
ax.set_yticks([20, 40, 60, 80, 100])
ax.set_yticklabels([])
ax.grid(color="lightgray", linestyle="dashed", alpha=0.6)

ax.set_xticks(angles + width / 2)
ax.set_xticklabels(metrics, fontsize=10)

for angle in [0, 3, 6, 9]:
    ax.plot(
        [angle * width, angle * width],
        [0, 100],
        color="black",
        linewidth=2
    )

ax.text(0.05, 1.05, "Attack", transform=ax.transAxes,
        color=colors["Attack"], fontsize=12, fontweight="bold")

ax.text(0.85, 1.05, "Defence", transform=ax.transAxes,
        color=colors["Defence"], fontsize=12, fontweight="bold")

ax.text(0.85, -0.05, "Possession", transform=ax.transAxes,
        color=colors["Possession"], fontsize=12, fontweight="bold")

ax.text(0.05, -0.05, "Progression", transform=ax.transAxes,
        color=colors["Progression"], fontsize=12, fontweight="bold")

plt.show()
```
